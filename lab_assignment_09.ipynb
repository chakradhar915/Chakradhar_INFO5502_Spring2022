```python
# You answer here
'''
1)we can perform the Z test as the we have above 30 samples
  Hypothesis Statement
  null hypothesis:P=0.05
  alternative hypothesis:P>0.05

2)
The normal test (sample proportion test) is the appropriate test.
Conditions:
* We need the sample size 
* We must know the SD(standard deviation) as well
* We must know the propotion 

3)
*pvalue = 46/384 #probability of given population
*p = 0.05 #This can be taken by 5%
*x = m.sqrt(p*(1-p))
*z = (pvalue-p)/(x/m.sqrt(384))
*z = 6.27512493759735
as z is greater than 1.96 Null hyphothesis rejected , Alternative hyphothesis accepted
'''
```

```python
import pandas as pd
import math as m

pvalue = 46/384 #probability of given population
p = 0.05 #This can be taken by 5%
x = m.sqrt(p*(1-p))


z = (pvalue-p)/(x/m.sqrt(384))
print("Z-state", z)


if z>1.96:
  print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
  print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```

```python
 #2(2-3)
def sim_and_test_stat(table, label, group_label):
  shuffled_column = table.sample(with_replacement = False).column(group_label)
  new_table = table.drop(group_label).with_column("shuffled label", shuffled_column)
  return find_stat(new_table, label, "shuffled label")

sim_and_test_stat(change_in_death_rates, "Murder Rate", "Death Penalty")
```
