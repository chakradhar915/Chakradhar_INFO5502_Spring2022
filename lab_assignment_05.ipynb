Question 3 (15 points). In this question, you are required to create visualization for time series data. 

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
!pip install yake
```

```python
# Your python code here:

import pandas as pd
import matplotlib.pyplot as plt
import yake
df_data = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv", sep=",")
# we drop the rows that have nan values in year
df_data = df_data[df_data['year'].isna()!=True]
# we type cast the year column
df_data['year'] = df_data['year'].astype('int64')
```

```python
df_data.info()
#we can see we have no null values for all atrributes.
```

```python
for year in df_data['year'].unique():
    print("The Top 5 keywords of the year {} are following:".format(year))
    numOfKeywords = 5
    text = " ".join(rawdata for rawdata in df_data[df_data['year']==year]['title'].values)
    #extracting the top words by using the yake
    keywords_extract = yake.KeywordExtractor(lan="en", top=numOfKeywords, features=None)
    #taking it to list
    keywords = keywords_extract.extract_keywords(text)
    Score = [y for x, y in keywords]
    Keyword=[x for x, y in keywords]
    plt.ylim(0, 0.01)
    plt.bar(Keyword,Score,label=year,color=['black', 'red', 'green', 'blue', 'cyan'])
    plt.xticks(rotation=90)
    for kw in keywords:
        print(kw[0])
    print("\n")
```
